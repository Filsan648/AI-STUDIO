```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/ncr_ride_bookings.csv')
```

**Column Descriptions**

The dataset contains several columns. Below is the meaning and purpose of each column:

**Date**:Date of the booking


**Time** :Time of the booking

**Booking** :ID	Unique identifier for each ride booking

**Booking Status**: Status of booking (Completed, Cancelled by Customer, Cancelled by Driver, etc.)

**Customer ID**:	Unique identifier for customers

**Vehicle Type**	:Type of vehicle (Go Mini, Go Sedan, Auto, eBike/Bike, UberXL, Premier Sedan)

**Pickup Location**	Starting location of the ride

**Drop Location**:	Destination location of the ride

**Avg VTAT	Average**: time for driver to reach pickup location (in minutes)

**Avg CTAT**:	Average trip duration from pickup to destination (in minutes)

**Cancelled Rides by Customer**:	Customer-initiated cancellation flag

**Reason for cancelling by Customer**:	Reason for customer cancellation
**Cancelled Rides by Driver**:	Driver-initiated cancellation flag

**Driver Cancellation Reason**	:Reason for driver cancellation

**Incomplete Rides**:	Incomplete ride flag

**Incomplete Rides Reason**:	Reason for incomplete rides
**Booking Value**:Total fare amount for the ride

**Ride Distance**:	Distance covered during the ride (in km)
**Driver Ratings**:	Rating given to driver (1-5 scale)

**Customer Rating**:	Rating given by customer (1-5 scale)
**Payment Method**:	Method used for payment (UPI, Cash, Credit Card, Uber Wallet, Debit Card)

**Data Cleaning**

```python
df["Booking Value"].unique()
df["Booking Value"]=df["Booking Value"].astype(float)
df["Booking Value"].fillna(df["Booking Value"].mean(),inplace=True)
df["Booking Value"]
```

```python
df["Ride Distance"].unique()
df["Ride Distance"]=df["Ride Distance"].astype(float)
df["Ride Distance"]=df["Ride Distance"].fillna(df["Ride Distance"].mean(),inplace=True)
```

```python
df["Driver Ratings"]=df["Driver Ratings"].astype(float)
df["Driver Ratings"]=df["Driver Ratings"].fillna(df["Driver Ratings"].mean(),inplace=True)
```

```python
df["Customer Rating"]=df["Customer Rating"].astype(float)
df["Customer Rating"]=df["Customer Rating"].fillna(df["Customer Rating"].mean(),inplace=True)
```

```python
boking_percent=[]
boking_percent.append( (df.loc[df["Booking Status"]=="Incomplete"].shape[0]/df["Booking Status"].count())*100)
boking_percent.append( (df.loc[df["Booking Status"]=="Completed"].shape[0]/df["Booking Status"].count())*100)
plt.title('Booking Status Distribution')
plt.pie(boking_percent,labels=["Incomplete","Completed"],autopct='%1.1f%%')
```

```python
vehicule_type=[]
vehicule_type.append(df.loc[df["Vehicle Type"]=='Go Sedan'].shape[0])
vehicule_type.append(df.loc[df["Vehicle Type"]=='Auto'].shape[0])
vehicule_type.append(df.loc[df["Vehicle Type"]=='Premier Sedan'].shape[0])
vehicule_type.append(df.loc[df["Vehicle Type"]=='Bike'].shape[0])
vehicule_type.append(df.loc[df["Vehicle Type"]=='Go Mini'].shape[0])
vehicule_type.append(df.loc[df["Vehicle Type"]=='Uber XL'].shape[0])
vehicule_type.append(df.loc[df["Vehicle Type"]=='eBike'].shape[0])
vehicule_type
plt.bar(['Go Sedan', 'Auto', 'Premier Sedan', 'Bike', 'Go Mini', 'Uber XL',
       'eBike'],vehicule_type,width=0.2)
plt.show()
```

```python
cancelled_Bydreiver = df.loc[df["Cancelled Rides by Driver"] ==1.0].shape[0]
casencelled_ByCustomer=df.loc[df["Cancelled Rides by Customer"] ==1.0].shape[0]
Total_Cancelled=casencelled_ByCustomer+casencelled_ByCustomer
plt.pie([(cancelled_Bydreiver/Total_Cancelled)*100,(casencelled_ByCustomer/Total_Cancelled)*100],labels=["cancelled By driver","Cancelled By Customer"],autopct='%1.1f%%')
```

```python
Reason_Cancelled_ByDriver=df.groupby(df["Driver Cancellation Reason"]).size()

Reason_Cancelled_ByDriver.plot(kind="bar", figsize=(8,5))
```

```python
reason_cancelled_by_customer=df.groupby(df["Reason for cancelling by Customer"]).size()
reason_cancelled_by_customer.plot(kind="bar", figsize=(8,5))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].boxplot(df["Avg CTAT"].dropna(), vert=True, patch_artist=True)
axes[0].set_title("Durée moyenne des trajets (Avg CTAT)")
axes[0].set_ylabel("Durée en minutes")
axes[1].boxplot(df["Avg VTAT"].dropna(), vert=True, patch_artist=True)
axes[1].set_title("Temps moyen avant prise en charge (Avg VTAT)")
axes[1].set_ylabel("Durée en minutes")
plt.tight_layout()
plt.show()
```

```python
import datetime
df["Time"] = pd.to_datetime(df["Time"], format="%H:%M:%S")

morning_rides = df.loc[
    (df["Time"].dt.time >= datetime.time(1, 0, 0)) &
    (df["Time"].dt.time < datetime.time(12, 0, 0))
].shape[0]

Nonne = df.loc[
    (df["Time"].dt.time >= datetime.time(12, 0, 0)) &
    (df["Time"].dt.time < datetime.time(15, 0, 0))
].shape[0]

Afternone_rides = df.loc[
    (df["Time"].dt.time >= datetime.time(15, 0, 0)) &
    (df["Time"].dt.time < datetime.time(18, 0, 0))
].shape[0]

Night_rides = df.loc[
    (df["Time"].dt.time >= datetime.time(18, 0, 0)) |
    (df["Time"].dt.time < datetime.time(0, 0, 0))
].shape[0]

Midnight_rides = df.loc[
    (df["Time"].dt.time >= datetime.time(0, 0, 0)) &
    (df["Time"].dt.time < datetime.time(1, 0, 0))
].shape[0]
```

```python
plt.bar(["morning_rides","Nonne","Afternone_rides","Night_rides","Midnight_rides"],[morning_rides,Nonne,Afternone_rides,Night_rides,Midnight_rides])
```
